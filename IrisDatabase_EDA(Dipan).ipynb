# **Line Plot:**

**Plot a line graph showing the sepal length of the three species of Iris flowers. Use
different colors to represent each species.**

```python
data['Species'].value_counts()
```

```python
setosa_length = data[data['Species']=='Iris-setosa']['SepalLengthCm']
versicolor_length = data[data['Species']=='Iris-versicolor']['SepalLengthCm']
virginica_length = data[data['Species']=='Iris-virginica']['SepalLengthCm']

x = range(1, len(setosa_length)+1)

plt.figure(figsize=(8,6))

plt.plot(x, setosa_length, label='Setosa', color='blue')
plt.plot(x, versicolor_length, label='Versicolor', color='green')
plt.plot(x, virginica_length, label='Virginica', color='red')

plt.xlabel('Sample')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length of Iris Species')
plt.legend()

plt.grid(True)
plt.show()
```

# **Scatter Plot:**

**Create a scatter plot showing the relationship between the petal length and petal
width. Use different colors to represent each species.**

```python
color_data = {
    'Iris-setosa': 'blue',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'red'
}

plt.figure(figsize=(6,4))

for species in data['Species'].unique():
  subset= data[data['Species'] == species]
  plt.scatter(subset['PetalLengthCm'], subset['PetalWidthCm'], color=color_data[species], label=species)

plt.xlabel('Petal Length(cm)', fontsize=12)
plt.ylabel('Petal Width(cm)', fontsize=12)
plt.title('Scatterb Plot of Petal Length vs. Petal Width by Species', fontsize=14)
plt.legend()
plt.show()
```

# **Histogram:**

**Plot a histogram showing the distribution of the petal widths. Use different colors to
represent each species and display a legend.**

```python
color_map = {
    'Iris-setosa': 'blue',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'red'
}

setosa_width = data[data['Species']=='Iris-setosa']['PetalWidthCm']
versicolor_width = data[data['Species']=='Iris-versicolor']['PetalWidthCm']
virginica_width= data[data['Species']=='Iris-virginica']['PetalWidthCm']

plt.figure(figsize=(6,5))

plt.hist([setosa_width,versicolor_width,virginica_width], bins=30, histtype='bar', color=[color_map[s] for s in ['Iris-setosa','Iris-versicolor','Iris-virginica']], label=['Iris-setosa','Iris-versicolor','Iris-virginica'])

plt.xlabel('Petal Width (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Petal Widths by Species', fontsize=14)
plt.legend()

plt.grid(True)
plt.show()
```

# **Bar Plot:**

**Create a bar plot showing the average sepal width for each species.**

```python
species = data['Species'].unique()
sepal_width = []

for s in species:
  subset = data[data['Species'] == s]
  avg_width = subset['SepalWidthCm'].mean()
  sepal_width.append(avg_width)

plt.figure(figsize=(6,4))
plt.bar(species, sepal_width, color=['purple','navy','crimson'])

plt.xlabel('Species', fontsize=12)
plt.ylabel('Average Sepal Width (cm)', fontsize=12)
plt.title('Average Sepal Width by Species', fontsize=14)
plt.show()
```

# **Pie Chart:**

**Assume that the dataset represents the entire population of Iris flowers. Create a pie
chart showing the proportion of each species in the population.**

```python
species_count = data['Species'].value_counts()
labels = species_count.index.to_list()
colors = ['palegreen','bisque','plum']

plt.figure(figsize=(6,4))
plt.pie(species_count, colors=colors, autopct='%1.1f%%', labels=labels)

plt.title('Proportion of Iris Species in Population', fontsize=14)
plt.show()
```

# **Box Plot:**

**Create a box plot showing the distribution of sepal length for each species.**

```python
plt.figure(figsize=(6,4))

plt.boxplot([setosa_length, versicolor_length, virginica_length], labels=['Iris-setosa','Iris-versicolor','Iris-virginica'], patch_artist=True)

plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.title('Distribution of Sepal Length by Species', fontsize=14)
plt.show()
```
